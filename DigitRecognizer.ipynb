```python
import numpy as np

from sklearn.cluster import KMeans

import matplotlib.pyplot as plt

from sklearn import neighbors, datasets

import pandas as pd
```

```python
train = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/train.csv')
test = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/test.csv')
train.head()
```

```python
x_train = train.drop(labels = 'label', axis = 1)
y_train = train.label
```

```python
#Normalize Data for images
x_train = x_train / 255.0
test = test / 255.0

print('x_train', x_train.shape)
print('test', test.shape)
```

```python
clf = neighbors.KNeighborsClassifier(n_neighbors = 10, p = 2, weights = 'distance')
clf.fit(x_train, y_train)
y_pred = clf.predict(test)
```

```python
imgid = np.array(range(0,test.shape[0])).astype(int)
label = pd.Series(imgid,name='ImageId').astype(int)
prd = pd.Series(y_pred,name='Label').astype(int)
results = pd.concat([label, prd], axis = 1)
results.to_csv('digitRecognizer.csv', index = False)
```
